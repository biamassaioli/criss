```python
import pandas as pd
from seaborn import displot
import matplotlib.pyplot as plt
```

```python
caminho_arquivo = "/home/prietoab/sussuarana/clientes/davi/imagens/GAPED_imagens/"
nome_arquivo = "imagens.csv"

imagens = pd.read_csv(
    caminho_arquivo+nome_arquivo, 
    sep=";",
    decimal=","
)
```
